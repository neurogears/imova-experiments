```python
import os
import pandas as pd

data_dir = r"C:\Data\IMOVA\toydataset\imova\microphone\CMD000\2025-10-14T17-01-09"
datasets = {}

for file in os.listdir(data_dir):
    file_path = os.path.join(data_dir, file)
    if file.endswith('.csv') and file != 'Instructions.csv':
        datasets[file] = pd.read_csv(file_path)
    elif file.endswith('.xlsx'):
        datasets[file] = pd.read_excel(file_path)
    # Add more file type handlers if needed

print(f"Loaded datasets: {list(datasets.keys())}")
```
